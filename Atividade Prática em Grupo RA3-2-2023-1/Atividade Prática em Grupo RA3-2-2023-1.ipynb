# Exercício de integrais

Considerando as integrais a seguir você deverá plotar o gráfico de cada uma das equações que estão
sendo integradas e marcar a área entre os limites de integração. Depois, usando apenas álgebra,
calcular a área sobre estas curvas e, finalmente, verificar seus cálculos calculando a área sobre estas
curvas usando a Soma de Reimann. Esta Tarefa deve ser realizada no Google Colab usando o Python,
limitado ao uso das bibliotecas numpy, sympy e matplotlib.

```python
import matplotlib.pyplot as pyplot
import numpy as np
import sympy as sp
```

```python
def plot_int(f, x, min_x, max_x, label, tryEvalf=False):
  x_lim = 0.5

  # Calcula os valores das funções
  vals_x = np.linspace(min_x - x_lim, max_x + x_lim, 500)
  vals_f = np.array([f.subs(x, i) for i in vals_x], dtype=float)
  vals_x_i = np.linspace(min_x, max_x, 500)
  vals_f_i = np.array([f.subs(x, i) for i in vals_x_i], dtype=float)

  # Gráfico
  #pyplot.gca().set_aspect('equal', 'datalim')
  pyplot.xlabel(r"$x$")
  pyplot.ylabel(r"$y$")
  pyplot.axhline(y=0, linestyle="--", color="black")
  pyplot.axvline(x=min_x, linestyle="--", color="blue")
  pyplot.axvline(x=max_x, linestyle="--", color="blue")

  pyplot.plot(vals_x, vals_f, color="red", label=label)
  pyplot.fill_between(vals_x_i, vals_f_i, step="pre", alpha=0.4)

  # Mostra resultados
  area = sp.integrate(f, (x, min_x, max_x))
  if not tryEvalf:
    print(f"Solução: ({area})\n")
  else:
    print(f"Solução: ({area}) -> ~{area.evalf()}\n")
  pyplot.legend()
  pyplot.show()
```

```python
def calc_reimman(f, x, min_x, max_x, precision=1000):
  sum = 0
  delta_x = (max_x - min_x) / precision

  for i in range(precision):
    sum += delta_x * f.subs(x, delta_x * i + min_x)

  print(f"Integral pela aproximação Soma de Reimman: ({sum})")
```

a)$\int_{2.5}^{7} (40x^3+12x^2-9x) dx$

$$\int (40x^3+12x^2-9x) dx$$
$$\int (40x^3)dx + \int(12x^2)dx - \int(9x) dx$$
$$ 40\frac{1}{4}x^4+c + 12\frac{1}{3}x^3+c - 9\frac{1}{2}x^2+c$$
$$ \therefore F(x) = 10x^4 + 4x^3 - \frac{9}{2}x^2 + c$$

$$ F(7)-F(2.5) $$
$$ (24010 + 1372 - 220.5 + c )- (390.625 + 62.5 - 28.125 + c) $$
$$ 25161.5 - 425 +c$$
$$\therefore 24736.5+c$$

```python
#A
x = sp.Symbol('x')
f = 40*x**3 + 12*x**2 - 9*x
min_x = 2.5
max_x = 7

calc_reimman(f, x, min_x, max_x)
plot_int(f, x, min_x, max_x, r"$40x^3+12x^2-9x$")
```

```python
#B
t = sp.Symbol('t')
f = 12*t**7 - t**2 - t + 3
min_x = 7.5
max_x = 10

calc_reimman(f, t, min_x, max_x)
plot_int(f, t, min_x, max_x, r"$12t^7-t^2-t+3$")
```

```python
#C
x = sp.Symbol('x')
f = (2*x**5 - x + 3) / (x**2)
min_x = 1
max_x = 2

calc_reimman(f, x, min_x, max_x)
plot_int(f, x, min_x, max_x, r"$\frac{2x^5-x+3}{x^2}$", tryEvalf=True)
```
