```python
import matplotlib.pyplot as pyplot
import numpy as np

from mpmath import mp
mp.dps = 20


pyplot.title("Gráfico de $f\'(x)$ para $f(x)=2x+\sin{(3x)}$")
pyplot.xlabel("$x$")
pyplot.ylabel("$y$")

x = np.linspace(-10, 10, 10000)
f = 2*x + np.sin(3*x) # Função f(x)
df = 2 + 3 * np.cos(3*x) # Derivada de f(x)

delta_x = 10**-6
df_fd = ((2*(x + delta_x) + np.sin(3*(x + delta_x))) - (2*(x) + np.sin(3*(x)))) / delta_x # Aproximação da derivada de f(x) pela Fórmula das Diferenças Finitas
df_pm = ((2*(x + delta_x/2) + np.sin(3*(x + delta_x/2))) - (2*(x - delta_x/2) + np.sin(3*(x - delta_x/2)))) / delta_x # Aproximação da derivada de f(x) pela Regra do Ponto Médio

pyplot.plot(x, f, label="$f(x)$")
pyplot.plot(x, df, label="$f\'(x)$")
pyplot.plot(x, df_fd, label="Aproximação de $f\'(x)$ com Diferenças Finitas")
pyplot.plot(x, df_pm, label="Aproximação de $f\'(x)$ com Ponto Médio")

pyplot.legend()
pyplot.show()


delta_x = mp.mpf(10**-6)
# Escolhi x = 1 como o valor em que a derivada é negativa
x_dfneg = mp.mpf(1)
dfneg = 2 + 3 * mp.cos(3*x_dfneg)
fd_dfneg = (((2*(x_dfneg + delta_x) + mp.sin(3*(x_dfneg + delta_x))) - (2*(x_dfneg) + mp.sin(3*(x_dfneg)))) / delta_x)
pm_dfneg = (((2*(x_dfneg + delta_x/2) + mp.sin(3*(x_dfneg + delta_x/2))) - (2*(x_dfneg - delta_x/2) + mp.sin(3*(x_dfneg - delta_x/2)))) / delta_x)
print(f"Erro com derivada negativa usando a aproximação com a Fórmula das Diferenças Finitas é de {fd_dfneg - dfneg}")
print(f"Erro com derivada negativa usando a aproximação com a Regra do Ponto Médio é de {pm_dfneg - dfneg}")


# Escolhi x = ~0.767 como o valor em que a derivada é nula
x_dfnula = mp.mpf(1)
dfnula = 2 + 3 * mp.cos(3*x_dfnula)
fd_dfnula = (((2*(x_dfnula + delta_x) + mp.sin(3*(x_dfnula + delta_x))) - (2*(x_dfnula) + mp.sin(3*(x_dfnula)))) / delta_x)
pm_dfnula = (((2*(x_dfnula + delta_x/2) + mp.sin(3*(x_dfnula + delta_x/2))) - (2*(x_dfnula - delta_x/2) + mp.sin(3*(x_dfnula - delta_x/2)))) / delta_x)
print(f"Erro com derivada nula usando a aproximação com a Fórmula das Diferenças Finitas é de {fd_dfnula - dfnula}")
print(f"Erro com derivada nula usando a aproximação com a Regra do Ponto Médio é de {pm_dfnula - dfnula}")


# OBS.: Dá para observar que o erro é constante!
```
