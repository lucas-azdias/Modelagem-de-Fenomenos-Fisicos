# Questão 1
Uma doença se espalha de forma exponencial, com uma taxa de contágio de 0,05 por dia. Se inicialmente 1000 pessoas são infectadas, faça um gráfico que mostre como o número de pessoas infectadas varia com o tempo e determine quantas pessoas estarão infectadas após 30 dias. Utilize a equação $I = I_0 * e^{rt}$, onde $I_0$ é o número inicial de infectados, `r` é a taxa de contágio diária (em decimal) e $t$ é o tempo decorrido em dias.



```python
import matplotlib.pyplot as pyplot
import numpy as np


t = np.linspace(0, 20)
i0 = 1000
r = 0.05

i = i0 * np.e**(r * t)

pyplot.xlabel("Tempo (dias)")
pyplot.ylabel("Número de infectados")
pyplot.plot(t, i)
pyplot.show()
```

# Questão 2
Usando, a seu critério, as bibliotecas do Python, crie um gráfico para cada uma das funções a seguir. No label do gráfico deverá estar a função impressa com qualidade matemática (latex) e uma label dizendo se a função é crescente ou decrescente. 

1. $f(x) = 3x^2 - 6^x +48$

2. $f(x) = \log_{10}(x^2 - 3x) - \log_{10}(x - 3)$

```python
import matplotlib.pyplot as pyplot
import numpy as np


x = np.linspace(0, 20)


f1 = 3*x**2 - 6**x + 48

pyplot.title(r'$f(x)=3x^{2}-6^{x}+48$')
pyplot.xlabel(r'$x$')
pyplot.ylabel(r'$f(x)$')
pyplot.plot(x, f1)
pyplot.gca().annotate('Decrescente', xy=(1, 0), xycoords='data', fontsize=10, va="bottom")
pyplot.show()


f2 = np.log10(x**2 - 3*x) - np.log10(x - 3)  # Em 3 e em 0, é indefinida pois gera log(0)

pyplot.title(r'$f(x)=\log_{10}(x^{2}-3x)-\log_{10}(x-3)$')
pyplot.xlabel(r'$x$')
pyplot.ylabel(r'$f(x)$')
pyplot.plot(x, f2)
pyplot.gca().annotate('Crescente', xy=(5, 0.5), xycoords='data', fontsize=10, va="bottom")
pyplot.show()
```

Usando o Sympy, ou qualquer outro recurso disponível na internet, escreva, com qualidade matemática o passo-a-passo algébrico que leva a determinação dos seguintes limites. Ao final, para cada uma das equações você deverá traçar um gráfico da função. 

1. $\lim_{x \to 1} \frac{x^{10} - 1}{x^5 - 1}$

2. $\lim_{x \to \pi} \frac{\cos(x) - 1}{x - \pi}$

OBS.: Usei os sites [AllMath](https://www.allmath.com/limit-calculator.php), [WolframAlpha](https://wolframalpha.com) e [Desmos](https://desmos.com)

1.  RESOLUÇÃO
$$\lim_{x \to 1} \frac{x^{10} - 1}{x^5 - 1}$$
$$\lim_{x \to 1} (\frac{(x-1)(x+1)(x^4-x^3+x^2-x+1)(x^4+x^3+x^2+x+1)}{(x-1)(x^4+x^3+x^2+x+1)})$$
$$\lim_{x \to 1} ((x+1)(x^4-x^3+x^2-x+1))$$
$$\text{Aplicando-se a substituição:}$$
$$((1+1)(1-1+1-1+1))$$
$$(2)(1)$$
$$\therefore2$$

2. RESOLUÇÃO
$$\lim_{x \to \pi} \frac{\cos(x) - 1}{x - \pi}$$
$$\text{Conforme }x\text{ tende a }\pi\text{, }cos(x) - 1\text{ tende a }-2$$
$$\lim_{x \to \pi} \frac{-2}{x - \pi}$$
$$\text{Conforme }x\text{ tende a }\pi\text{, }x-\pi\text{ tende a }0$$
$$\frac{-2}{0}$$
$$\therefore\text{Assim, o resultado tende a }\infty\text{,}$$
$$\text{sendo que }(\lim_{x \to \pi^{+}}\frac{\cos(x) - 1}{x - \pi})\to-\infty\text{ e }(\lim_{x \to \pi^{-}}\frac{\cos(x) - 1}{x - \pi})\to+\infty$$

```python
import sympy as sp


x = sp.Symbol('x')


f1 = (x**10 - 1) / (x**5 - 1)
sp.plot(f1)


f2 = (sp.cos(x) - 1) / (x - sp.pi)
sp.plot(f2, xlim=[-20, 20], ylim=[-10, 10])
```
