3. Usando o Python e as fun칞칫es trigonom칠tricas dispon칤veis nesta linguagem de programa칞칚o 
e  suas  bibliotecas  crie  uma  fun칞칚o  que  receba  um  칙ngulo,  em  graus,  e  plote  um  circulo 
trigonom칠trico de raio 1, centrado na origem destacando este 칙ngulo, seu seno, seu cosseno 
e sua tangente. 

```python
#3

import matplotlib.pyplot as pyplot
import numpy as np


def plotUnitCircle(angle):
  # Mostra o gr치fico do c칤rculo unit치rio com o 칙ngulo dado (em graus) e indica
  # o seno, o cosseno e a tangente dele
  chartsize = (10, 10) # Define o tamanho do gr치fico

  # Configura칞칚o inicial do gr치fico
  fig, ax = pyplot.subplots(figsize=chartsize)
  fig.canvas.manager.set_window_title('Gr치fico') # Define o nome da janela
  ax.set_title(f"Unit Circle with the angle {angle}춿") # Define o t칤tulo do gr치fico
  ax.set_aspect('equal', 'datalim') # Obriga os eixos a serem proporcionais

  ax.axhline(y=0, linestyle='--', color='black') # Coloca o eixo X no gr치fico
  ax.axvline(x=0, linestyle='--', color='black') # Coloca o eixo Y no gr치fico
  ax.axvline(x=1, color='black') # Coloca o eixo da tangente no gr치fico

  angle_rad = angle * np.pi / 180 # Converte o 칙ngulo de graus para radianos

  alpha = np.linspace(0, 2 * np.pi, 500)
  ax.plot(np.cos(alpha), np.sin(alpha), color='blue') # Coloca um c칤rculo unit치rio na origem no gr치fico

  alpha = np.linspace(0, angle_rad, 500)
  ax.plot(0.1 * np.cos(alpha), 0.1 * np.sin(alpha), color='purple') # Coloca o semi-c칤rculo do 칙ngulo no gr치fico

  sin = np.sin(angle_rad)
  cos = np.cos(angle_rad)
  tan = np.tan(angle_rad)
  ax.plot(np.linspace(cos, cos), np.linspace(0, sin), color='green', label='Seno') # Coloca a linha do seno no gr치fico
  ax.plot(np.linspace(0, cos), np.linspace(0, 0), color='red', label='Cosseno') # Coloca a linha do cosseno no gr치fico
  ax.plot(np.linspace(0, 1), np.linspace(0, tan), color='orange', label='Tangente') # Coloca a linha da tangente no gr치fico
  ax.plot(np.linspace(0, cos), np.linspace(0, sin), linestyle='--', color='blue', label='Raio') # Coloca a linha do raio no gr치fico
  ax.plot(cos, sin, marker='o', color='blue') # Coloca o ponto na circunfer칡ncia pelo 칙ngulo no gr치fico
  ax.plot(1, tan, marker='o', color='orange') # Coloca o ponto na tangente pelo 칙ngulo no gr치fico
  ax.plot(0, 0, marker='o', color='black') # Coloca o ponto da origem no gr치fico

  # Anota칞칫es no gr치fico com indica칞칫es
  fig.gca().annotate('O', xy=(-0.075, -0.075), fontsize=10)
  fig.gca().annotate('Raio', xy=(cos / 2 - 0.05, sin / 2 + 0.05), fontsize=10, rotation=angle)
  fig.gca().annotate('Seno', xy=(cos + 0.025, sin / 3 + 0.05), fontsize=10, rotation=90)
  fig.gca().annotate('Cosseno', xy=(cos / 3,-0.075), fontsize=10)
  fig.gca().annotate('Tangente', xy=(2 / 3 - 0.05, 2 * tan / 3 + 0.05), fontsize=10, rotation=angle)

  ax.legend()
  pyplot.show()


# Exemplo para um 칙ngulo de 60춿
plotUnitCircle(angle=60)
```

4. Usando o Python e as fun칞칫es trigonom칠tricas dispon칤veis nesta linguagem de programa칞칚o 
e suas bibliotecas crie uma fun칞칚o que receba um n칰mero inteiro 洧녵 tal que 1곣洧녵곣5 e plote 
as  fun칞칫es  seno,  cosseno  e  tangente  para  todos  os  칙ngulos  poss칤veis  considerando  que 洧녵 
representa o n칰mero per칤odos completos.

```python
#4

import matplotlib.pyplot as pyplot
import matplotlib.widgets as widgets
import matplotlib.lines as lines
import numpy as np


def plotGraph(p):
  rangeP = [1, 5] # Define o intervalo para os per칤odos completos

  # Garantir que p est치 no intervalo adequado
  if p < min(rangeP):
    p = min(rangeP)
  elif p > max(rangeP):
    p = max(rangeP)

  chartsize = (6.5, 6.5) # Define o tamanho do gr치fico

  # Configura칞칚o inicial do gr치fico
  fig, ax = pyplot.subplots(figsize=chartsize)
  fig.canvas.manager.set_window_title('Gr치fico') # Define o nome da janela
  ax.set_title(f"Gr치fico das fun칞칫es trigonom칠tricas (com n={p})") # Define o t칤tulo do gr치fico
  ax.set_aspect('equal', 'datalim') # Obriga os eixos a serem proporcionais

  ax.axhline(y=0, linestyle='--', color='black') # Coloca o eixo X no gr치fico
  ax.axvline(x=0, linestyle='--', color='black') # Coloca o eixo Y no gr치fico

  x = np.linspace((p - 1) * np.pi * 2, p * np.pi * 2) # Define as coordenadas no eixo X
  sin = np.sin(x) # Define as coordenadas no eixo Y para o seno
  cos = np.cos(x) # Define as coordenadas no eixo Y para o cosseno
  tan = np.tan(x) # Define as coordenadas no eixo Y para o tangente

  ax.plot(x, sin, color='green', scaley=False) # Cria a fun칞칚o seno
  ax.plot(x, cos, color='orange', scaley=False) # Cria a fun칞칚o cosseno
  ax.plot(x, tan, color='red', scaley=False) # Cria a fun칞칚o tangente
  # OBS.: scaley=False foi usado para melhorar a visibilidade

  pyplot.show()


# Exemplo com 2 per칤odos completos
plotGraph(2)
```
