Considerando as funções listadas a seguir faça: **(a)** o gráfico das três funções; **(b)** o passo a passo algébrico, em latex, para encontrar o ângulo de inclinação da reta tangente no ponto $x=5$ para cada das funções:

1. $f(x) = (x^2 - 4)(x^3 - 8x-1)$
$$\frac{d}{dx}f(x)$$
$$\frac{d}{dx}[(x^2 - 4)(x^3 - 8x-1)]$$
$$(x^2 - 4)[\frac{d}{dx}(x^3 - 8x-1)]+(x^3 - 8x-1)[\frac{d}{dx}(x^2 - 4)]$$
$$(x^2 - 4)(3x^2 - 8)+(x^3 - 8x-1)(2x)$$
$$(x^4-20x^2+32)+(2x^4-16x^2-2x)$$
$$\therefore 3x^4-36x^2-2x+32$$

$$\text{Obtendo a inclinação para }x=5\text{:}$$
$$\begin{cases}
x=5\\
\frac{d}{dx}f(x)=3x^4-36x^2-2x+32
\end{cases}$$
$$\frac{d}{dx}f(5)=3(5)^4-36(5)^2-2(5)+32$$
$$\frac{d}{dx}f(5)=1875-900-10+32$$
$$\therefore\frac{d}{dx}f(5)=997$$
$$\text{Passando para ângulo:}$$
$$\alpha=\tan^{-1}(997)$$
$$\therefore\alpha\approx1,5697933181$$

2. $f(x) = \sin(x^3)\cos(x)$
$$\frac{d}{dx}f(x)$$
$$\frac{d}{dx}(\sin(x^3)\cos(x))$$
$$\sin(x^3)(\frac{d}{dx}\cos(x))+\cos(x)(\frac{d}{dx}\sin(x^3))$$
$$\sin(x^3)(-\sin(x))+\cos(x)(3x^2\cos(x^3))$$
$$\therefore-\sin(x)\sin(x^3)+3x^2\cos(x)\cos(x^3)$$

$$\text{Obtendo a inclinação para }x=5\text{:}$$
$$\begin{cases}
x=5\\
\frac{d}{dx}f(x)=-\sin(x)\sin(x^3)+3x^2\cos(x)\cos(x^3)
\end{cases}$$
$$\frac{d}{dx}f(5)=-\sin(5)\sin((5)^3)+3(5)^2\cos(5)\cos((5)^3)$$
$$\frac{d}{dx}f(5)=-\sin(5)\sin(125)+75\cos(5)\cos(125)$$
$$\therefore\frac{d}{dx}f(5)=-\sin(5)\sin(125)+75\cos(5)\cos(125)$$
$$\text{Passando para ângulo:}$$
$$\alpha=\tan^{-1}(-\sin(5)\sin(125)+75\cos(5)\cos(125))$$
$$\therefore\alpha\approx1,50902301909$$

3. $f(x) = x^2e^{-2x}\ln(x)$
$$\frac{d}{dx}f(x)$$
$$\frac{d}{dx}(x^2e^{-2x}\ln(x))$$
$$e^{-2x}\ln(x)(\frac{d}{dx}x^2)+x^2\ln(x)(\frac{d}{dx}e^{-2x})+x^2e^{-2x}(\frac{d}{dx}\ln(x))$$
$$e^{-2x}\ln(x)(2x)+x^2\ln(x)(-2(e^{-2x}))+x^2e^{-2x}(\frac{1}{x})$$
$$2xe^{-2x}\ln(x)+-2x^2e^{-2x}\ln(x)+xe^{-2x}$$
$$\therefore e^{-2x}(2x\ln(x)+-2x^2\ln(x)+x)$$

$$\text{Obtendo a inclinação para }x=5\text{:}$$
$$\begin{cases}
x=5\\
\frac{d}{dx}f(x)=e^{-2x}(2x\ln(x)+-2x^2\ln(x)+x)
\end{cases}$$
$$\frac{d}{dx}f(5)=e^{-2(5)}(2(5)\ln(5)+-2(5)^2\ln(5)+(5))$$
$$\therefore\frac{d}{dx}f(5)=e^{-10}(10\ln(5)+-100\ln(5)+5)$$
$$\text{Passando para ângulo:}$$
$$\alpha=\tan^{-1}(e^{-10}(10\ln(5)+-100\ln(5)+5))$$
$$\therefore\alpha\approx−0,00634906817443$$

```python
import sympy as sp
from math import e


x = sp.Symbol('x')

# Função 1
f1 = (x**2 - 4) * (x**3 - 8*x - 1)

# Função 2
f2 = sp.sin(x**3) * sp.cos(x)

# Função 3
f3 = x**2 * e**(-2*x) * sp.ln(x)

# Gráficos
sp.plot(f1, title="Função 1")
sp.plot(f2, title="Função 2")
sp.plot(f3, title="Função 3")
```

# Questão 2 (vale 2 pontos)
Uma empresa espacial está tentando descobrir o ponto mínimo da órbita de um satélite cuja função orbital é dada por: 
$$f(x)=7\sin(x)-2\cos(x)$$

Encontre o primeiro ponto de mínimo no intervalo $-10<=x<=10$.

Como sempre, você pode usar qualquer ferramenta disponível na internet para lhe auxiliar a resolver este problema. Lembre-se, contudo, que o sympy é uma boa alternativa. 

Em todas as questões você pode usar apenas os métodos, regras, técnicas e algoritmos que utilizamos na nossa disciplina, em sala, ou no material de apoio. Lembrem-se estamos limitados as bibliotecas sympy, numpy e matplotlib.

Para resolver esta questão você precisará de, no minimo, 4 casas decimais de precisão depois da vírgula. 

$f(x)=7\sin(x)-2\cos(x)$
$$\frac{d}{dx}f(x)$$
$$\frac{d}{dx}(7\sin(x)-2\cos(x))$$
$$(\frac{d}{dx}(7\sin(x))-\frac{d}{dx}(2\cos(x)))$$
$$(7\cos(x)-(-2\sin(x)))$$
$$\therefore7\cos(x)+2\sin(x)$$

Resolvendo para $\frac{d}{dx}f(x)=0$
$$7\cos(x)+2\sin(x)=0$$
$$7\cos(x)=-2\sin(x)$$
$$7=-2\frac{\sin(x)}{\cos(x)}$$
$$\tan(x)=-\frac{7}{2}$$
$$\therefore x=-\tan^{-1}(\frac{7}{2})$$

```python
import sympy as sp
import numpy as np


x = sp.Symbol('x')
f = 7*sp.sin(x) - 2*sp.cos(x)

x_res = float(-np.arctan(7/2))
y_res = f.subs(x, x_res)
p = np.pi * 2  # Período da função (fica padrão pois não há valor alterando x dentro das funções trigonométricas)

# Obtém a primeira após -10
while (x_res - p > -10):
  x_res -= p
# Não precisa atualizar y_res, pois é igual para todos os mínimos

print((x_res, y_res))
```
